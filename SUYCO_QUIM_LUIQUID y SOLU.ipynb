##Fuerzas Intermoleculares

###--------------------------------------------------------------------------

image.png

**Interacción dipolo-dipolo**: Se da entre moléculas que tienen dipolos permanentes. Un ejemplo de esto lo tenemos en la interacción entre moléculas de cloruro de hidrógeno (HCl) en estado líquido. En el HCl, el extremo de hidrógeno es ligeramente positivo, mientras que el extremo de cloro es ligeramente negativo. Estos dipolos hacen que las moléculas de HCl se atraigan entre sí.

**Interacción dipolo-dipolo inducido**: Se da entre moléculas que, aunque no tienen dipolos permanentes, pueden polarizarse en presencia de un dipolo cercano. Un ejemplo es el enlace entre moléculas de hidrógeno (H2) y cloruro de hidrógeno (HCl). El átomo de hidrógeno del H2, al acercarse a la molécula de HCl, induce una ligera polarización en la molécula de HCl, lo que genera una atracción entre el extremo positivo de HCl y el extremo negativo del H2.

**Interacción ion-dipolo:** Un ejemplo es la disolución de cloruro de sodio (NaCl) en agua. Cuando se disuelve NaCl, los iones Na+ y Cl- interactúan con las moléculas de agua, formando una capa de solvatación alrededor de cada ion.
image.png

**Fuerzas de dispersión (o de London):** Estas fuerzas actúan entre moléculas no polares y son causadas por fluctuaciones momentáneas en la distribución de electrones. Un ejemplo es la interacción entre moléculas de yodo (I2) en estado sólido, donde las moléculas de yodo no polar interactúan a través de fuerzas de dispersión.

image.png

El proceso de disolución de un sólido en un solvente es sencillo, para que esto se dé necesitamos que las fuerzas intermoleculares presentes en el soluto y solvente se rompan para así formas nuevas fuerzas intermoleculares entre ellas, obteniendo por consiguiente una mezcla homogénea.

Un ejemplo de ello lo tenemos cuando disolvemos la sal en agua, aquí necesariamente debe hacer presencia un agente externo, para este ejemplo la agitación del recipiente, esto originara un aumento en la energía cinética de las partículas, esto originara en un primero momento que las partículas del soluto se separen uniéndose con las moléculas de agua que se organizan alrededor de los iones de la sal debido a la polaridad del agua y la atracción entre los iones cargados y las moléculas de agua.

**¿Qué es la solvatación?**
La solvatación en cuando las moléculas del solvente rodean a la del soluto o iones de estos, formándose a la ya conocida “capa de solvatación”
Factores que influyen en el grado de solvatación:

T**emperatura**, esto afecta a la energia cinetica de las particulas originando que algunas fuerzas intermoleculares se rompan
***Electronegatividad, ***tanto del soluto como del solvente

###-----------------------------------------------------------------------------

image.png

**Proceso exotérmico:** cuando un soluto se disuelve en el solvente, las fuerzas intermoleculares del soluto y solvente se deben romper para formar nuevas fuerzas, pero ahora entre el soluto y solvente. Estas interacciones son lo suficientemente fuertes como para liberar energía en forma de calor al sistema. Un ejemplo común es la disolución de NaCl en agua, donde el agua se calienta cuando el NaCl se disuelve.

**Proceso endotérmico:** Las moléculas del soluto y el solvente se separan, pero para lograrlo, necesitan tomar prestada energía del entorno y al mismo tiempo generar fuerzas entre ellas., por eso, el agua tiende a enfriarse. Un ejemplo común de un proceso endotérmico es la disolución de nitrato de amonio en agua, donde el agua se enfría mientras el nitrato de amonio se disuelve.


###-----------------------------------------------------------------------------

image.png

Los factores que afectan de gran manera la solubilidad de un sólido a un líquido son: temperatura, superficie de contacto y agitación.

**Temperatura,** un aumento de la temperatura aumentará la energía cinética de las moléculas del sólido, favoreciendo la interacción entre las partículas del solvente con las del soluto haciéndolas más rápidas, en consecuencia, afectando la solubilidad.

**Superficie de contacto,** entre más pequeña las partículas del soluto ocurren dos cosas:
-	Mayor oportunidad para que las partículas del soluto interaccionen con las del solvente afectando de gran manera a la solubilidad.
-	Aumento en la energía cinética debido a que existe una mayor interacción entre las partículas del soluto y solvente, originando que las propias fuerzas intermoleculares entre los solutos y solventes se rompan fácilmente generando fuerzas intermoleculares entre ellas., convirtiéndose en una solución homogénea.

**Agitación,**al agitar el sistema, se aumentará la energía cinética de las partículas, lo que provoca una mayor colisión y choque entre las moléculas del soluto y solvente. obteniendo que las propias fuerzas intermoleculares entre los solutos y solventes se rompan fácilmente generando fuerzas intermoleculares entre ellas.


##Unidades de concentracion

image.png

**Porcentaje en masa:**
 El porcentaje en masa es la relación de la masa del soluto al peso total de la solución o mezcla, expresado en porcentaje.
 OJO: Ambos deben estar expresadas en las mismas unidades

 **Fracción molar:**
La fracción molar es la relación entre la cantidad de moles (soluto o solvente) y el total de moles de la solución.

**Molaridad**
La molaridad es la concentración de una solución expresada en moles del soluto por litro de solución.

**Molalidad**
La molalidad es la concentración de una solución expresada en moles de soluto por kilogramo de solvente.

***Ventajas:*** Es mejor obtener como dato el porcentaje en masa, ya que se puede expresar en kg o en gr, esto debido a que es adimensional , las unidades tanto en numerador como denominador son las mismas.
