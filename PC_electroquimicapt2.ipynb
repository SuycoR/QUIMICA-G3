```python
def anocato(r,p):
  if r < p:
    # Anodo = 0
    pos = 0
    return pos
  elif p < r:
    # Catodo = 1
    pos = 1
    return pos
```

```python
def fem_celda(p1,p2,fem1,fem2):

  if p1 == 1:
    femcelda = fem1 - fem2
  else:
    femcelda = fem2 - fem1

  return femcelda
```

```python
def pot_est(k,cr,cp):
  #Hallar el potencial estandar de cada uno
  #Diccionario de potenciales estandar para cada elemento según el cuadro adjunto anteriormente
  potenciales_reduccion = {'F':2.87, 'Co': [1.82,-0.28], 'Ce':1.61, 'Au':1.5, 'Cl':1.36, 'Br':1.07, 'Hg':[0.92,0.85], 'Ag':0.80, 'I':0.53, 'Cu':[0.34,0.15], 'Sn':[0.13,-0.14],'Pb':-0.13, 'Ni':-0.25, 'Cd':-0.4, 'Cr':-0.74, 'Zn':-0.76,'Mn':-1.18, 'Al':-1.66, 'Be':-1.85, 'Mg':-2.37, 'Na':-2.71, 'Ca':-2.87, 'Sr':-2.89, 'Ba':-2.90, 'K':-2.93, 'Li':-3.05,'Fe':[0.77,-0.44], }

  if type(potenciales_reduccion.get(k)) != list:
    potential = potenciales_reduccion.get(k)
  else:
    if k == 'Co':
      if cr == 3 and cp == 2:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[0]
      elif cr == 2 and cp == 0:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[1]
    elif k == 'Hg':
      if cr == 2 and cp == 2:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[0]
      elif cr == 2 and cp == 0:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[1]
    elif k == 'Fe':
      if cr == 3 and cp == 2:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[0]
      elif cr == 2 and cp == 0:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[1]
    elif k == 'Sn':
      if cr == 4 and cp == 2:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[0]
      elif cr == 2 and cp == 0:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[1]
    elif k == 'Cu':
      if cr == 2 and cp == 0:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[0]
      elif cr == 2 and cp == 1:
        potencial_list = potenciales_reduccion.get(k)
        potential = potencial_list[1]

  return potential
```

```python
def cambEl(fem):
  if fem < 0:
    print("El cambio de energía de Gibbs es positivo, es decir, el proceso es","\033[1m" + "forzado"+ "\033[0m","por lo tanto, estamos hablando de una","\033[1m" + "electrólisis"+ "\033[0m")
  else:
    print("El cambio de energía de Gibbs es negativo, es decir, el proceso es","\033[1m" + "espontáneo"+ "\033[0m","por lo tanto, estamos ante una","\033[1m" + "celda galvánica"+ "\033[0m")
```

```python
def inicio():
  print("Bienvenido, recuerda que estamos en un ambiente de 25° C:)")

  #Primer elemento
  elemento1 = input("Por favor, ingrese el primer elemento (su símbolo):")
    #Cargas del primer elemento
  cargareact1 = float(input(f"Ingrese su carga cuando {elemento1} está en los reactivos:"))
  cargaprod1 = float(input(f"Ingrese su carga cuando {elemento1} está en los productos:"))
    # ¿Ánodo o cátodo?
  posicion1 = anocato(cargareact1,cargaprod1)
    # Potencial estándar
  pestandar1 = pot_est(elemento1,cargareact1,cargaprod1)
  #Segundo elemento
  elemento2 = input("Por favor, ingrese el segundo elemento (su símbolo):")
    #Cargas del primer elemento
  cargareact2 = float(input(f"Ingrese su carga cuando {elemento2} está en los reactivos:"))
  cargaprod2 = float(input(f"Ingrese su carga cuando {elemento2} está en los productos:"))
    # ¿Ánodo o cátodo?
  posicion2 = anocato(cargareact2,cargaprod2)
    # Potencial estándar
  pestandar2 = pot_est(elemento2,cargareact2,cargaprod2)

  #Hallar el fem de la celda
  fem_celdan = fem_celda(posicion1,posicion2,pestandar1,pestandar2)
  print(f"El fem de la celda es {fem_celdan}V")
  cambEl(fem_celdan)

```
